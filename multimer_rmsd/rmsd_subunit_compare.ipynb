```python
%pylab inline
import os
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import pytraj as pt
```

```python
fig,(ax1,ax2,ax3,ax4,ax5)=subplots(1,5,figsize=(24,4))
E_name=['L1','L2','L3','L4','L5','L6','L7','L8']
for rep,ax in enumerate([ax1,ax2,ax3,ax4,ax5]):
    path="/Volumes/zhili_8TB/new_model/unact_rep0"+str(rep+1)+"/MMGBSA/"
    os.chdir(path)

    traj = pt.iterload('fitted_last_8L1.nc', 'L1.parm7',stride=1) #load the trajectory
    avg_pdb=pt.load('avg_L1.pdb','L1.parm7')   #load the avg structrue of L1
    pt.superpose(traj, ref=avg_pdb[0])    #impose on the avg structure

    for i in range(8):
        rmsd_data = pt.rmsd(traj, ref=traj[i])
        if i==0:
            rmsd_flow=rmsd_data
        else:
            rmsd_flow=np.vstack((rmsd_flow,rmsd_data))

    heatmap_E = pd.DataFrame(np.round(rmsd_flow,4),index=E_name,columns =E_name)
    sns.heatmap(heatmap_E,annot=True,ax=ax,vmax=5)
    ax.set_xlabel('rep'+str(rep+1))
plt.suptitle('rmsd of L1 comparison in apo 5 repeats')
```
